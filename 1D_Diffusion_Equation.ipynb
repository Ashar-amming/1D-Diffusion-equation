```python
import numpy
from matplotlib import pyplot
import time, sys
%matplotlib inline
```

```python
nx = 81
dx = 2/(nx-1)
nt = 20
nu = 0.3
sigma = 0.2
dt = sigma*dx**2/nu
```

```python
u = numpy.ones(nx)
u[int(0.5/dx):int(1/dx+1)] = 2

pyplot.plot(numpy.linspace(0,2,nx),u)
```

```python
un = numpy.ones(nx)

for n in range(nt):
    un = u.copy()
    for i in range(1,nx-1):
        u[i] = un[i] + nu*dt/dx**2  * (un[i+1] -2*u[i] + u[i-1])
```

```python
pyplot.plot(numpy.linspace(0,2,nx),u)
```
